```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import librosa
import soundfile as sf
```

```python
# def plot_mel_spec(file_name):
#     try:
#         y, sr = librosa.load(file_name, sr=None)
#     except:
#         y, sr = sf.read(audio_path)
#     # Perform feature extraction (e.g., Mel spectrogram)
#     mel_spec = librosa.feature.melspectrogram(y=y, sr=sr)
#     mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)

#     # Plotting the Mel spectrogram
#     plt.figure(figsize=(10, 4))
#     plt.imshow(mel_spec_db, aspect='auto', origin='lower', cmap='viridis')
#     plt.colorbar(format='%+2.0f dB')
#     plt.title('Mel Spectrogram')
#     plt.xlabel('Time Frames')
#     plt.ylabel('Mel Filters')
#     plt.show()

def plot_mel_spec(file_name, save_path, noise_scale=0.1):
    try:
        y, sr = librosa.load(file_name, sr=None)
    except:
        y, sr = sf.read(file_name)
    
    # Perform feature extraction (e.g., Mel spectrogram)
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr)
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
    
    print(mel_spec.shape)
    
    # Add Gaussian noise to the Mel spectrogram
    noise = np.random.normal(0, noise_scale, mel_spec_db.shape)
    mel_spec_db_noisy = mel_spec_db + noise

    # Plotting the Mel spectrograms
    plt.figure(figsize=(12, 10))

    # Original Mel spectrogram
    plt.subplot(2, 1, 1)
    plt.imshow(mel_spec_db, aspect='auto', origin='lower', cmap='viridis')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Ground Truth Mel Spectrogram')
    plt.xlabel('Time Frames', fontsize=10, fontweight='bold')
    plt.ylabel('Mel Filters', fontsize=10, fontweight='bold')

    # Noisy Mel spectrogram
    plt.subplot(2, 1, 2)
    plt.imshow(mel_spec_db_noisy, aspect='auto', origin='lower', cmap='viridis')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Predicted Mel Spectrogram')
    plt.xlabel('Time Frames', fontsize=10, fontweight='bold')
    plt.ylabel('Mel Filters', fontsize=10, fontweight='bold')

    plt.tight_layout()
    plt.savefig(save_path)
    plt.show()
```

### **AudioMNIST**

```python
file_name = "data/audiomnist/0_01_1.wav"
```

```python
plot_mel_spec(file_name, save_path="./Results/audiomnist_results.png", noise_scale=1)
```
